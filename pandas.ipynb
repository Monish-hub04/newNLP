```python
import pandas as pd

# Create Series
s = pd.Series(
    [10, 20, 30, 40, 50],
    index=["A", "B", "C", "D", "E"]
)

# Display Series
print(s)

# Access values
print("First Value =", s[0])

# Mean
print("Mean =", s.mean())

# Standard Deviation
print("Standard Deviation =", s.std())

# Maximum
print("Maximum =", s.max())

# Minimum
print("Minimum =", s.min())

# Filtering
print("Values Greater Than 20")
print(s[s > 20])
```

```python
import pandas as pd
import numpy as np

# Create dataset
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [20, 21, np.nan, 23, 24],
    "Marks": [90, 85, 88, np.nan, 95],
    "Department": ["IT", "CSE", "IT", "ECE", "CSE"]
}

df = pd.DataFrame(data)

# Display dataset
print(df)

# First rows
print(df.head())

# Last rows
print(df.tail())

# Dataset information
print(df.info())

# Statistical summary
print(df.describe())

# Mean
print(df.mean(numeric_only=True))

# Standard deviation
print(df.std(numeric_only=True))

# Missing values
print(df.isnull().sum())

# Fill missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)

df["Marks"].fillna(df["Marks"].mean(), inplace=True)

# Create new column
df["Bonus"] = df["Marks"] + 5

# GroupBy
print(df.groupby("Department").mean(numeric_only=True))

# Correlation
print(df.corr(numeric_only=True))

# Sorting
print(df.sort_values("Marks"))

# One hot encoding
print(pd.get_dummies(df["Department"]))

# Export CSV
# df.to_csv("output.csv")
```
