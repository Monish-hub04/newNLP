```python
import pandas as pd

from scipy.stats import zscore

# Dataset
data = {
    "Marks": [10, 12, 14, 15, 16, 18, 100]
}

df = pd.DataFrame(data)

# Original dataset
print("Original Dataset")
print(df)

# -------------------------
# Z-Score
# -------------------------

df["Z_score"] = zscore(
    df["Marks"]
)

print("\nZ-Score")
print(df)

# -------------------------
# Detect Z-score Outliers
# -------------------------

z_outliers = df[
    df["Z_score"].abs() > 3
]

print("\nZ-Score Outliers")
print(z_outliers)

# -------------------------
# Remove Z-score Outliers
# -------------------------

z_removed = df[
    df["Z_score"].abs() <= 3
]

print("\nAfter Removing Z-Score Outliers")
print(z_removed)

# -------------------------
# IQR Method
# -------------------------

Q1 = df["Marks"].quantile(0.25)

Q3 = df["Marks"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

# Detect IQR outliers
iqr_outliers = df[
    (df["Marks"] < lower) |
    (df["Marks"] > upper)
]

print("\nIQR Outliers")
print(iqr_outliers)

# Remove IQR outliers
iqr_removed = df[
    (df["Marks"] >= lower) &
    (df["Marks"] <= upper)
]

print("\nAfter Removing IQR Outliers")
print(iqr_removed)
```
