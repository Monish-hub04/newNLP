```python
import pandas as pd
import numpy as np

# Dataset
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [20, np.nan, 22, 21, np.nan],
    "Marks": [90, 85, np.nan, 88, 95],
    "Gender": ["Male", "Female", np.nan, "Male", "Female"]
}

df = pd.DataFrame(data)

# Display original dataset
print("Original Dataset")
print(df)

# Find missing values
print("\nMissing Values")
print(df.isnull().sum())

# Fill numerical columns using mean
df["Age"].fillna(
    df["Age"].mean(),
    inplace=True
)

df["Marks"].fillna(
    df["Marks"].mean(),
    inplace=True
)

# Fill categorical column using mode
df["Gender"].fillna(
    df["Gender"].mode()[0],
    inplace=True
)

# Create new column
df["Bonus_Marks"] = df["Marks"] + 5

# Display updated dataset
print("\nUpdated Dataset")
print(df)
```
