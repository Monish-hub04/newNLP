```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

# Dataset
data = {
    "Age": [20, 25, 30, 35, 40],
    "Income": [1000, 2000, 3000, 4000, 5000]
}

df = pd.DataFrame(data)

# Original dataset
print("Original Dataset")
print(df)

# -------------------------
# Min-Max Scaling
# -------------------------

minmax = MinMaxScaler()

df["Age_MinMax"] = minmax.fit_transform(
    df[["Age"]]
)

# -------------------------
# Standard Scaling
# -------------------------

standard = StandardScaler()

df["Income_Standard"] = standard.fit_transform(
    df[["Income"]]
)

# -------------------------
# Log Transformation
# -------------------------

df["Log_Income"] = np.log(
    df["Income"]
)

# Display updated dataset
print("\nTransformed Dataset")
print(df)
```
