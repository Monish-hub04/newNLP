```python
import numpy as np

from scipy.integrate import quad
from scipy.optimize import fsolve

# -------------------------
# Integral
# ∫ x^2 dx from 0 to 1
# -------------------------

def integral_function(x):
    return x**2

integral_result, error = quad(
    integral_function,
    0,
    1
)

print("Integral =", integral_result)

# -------------------------
# Linear Equation
# 2x + 3 = 7
# -------------------------

def linear_equation(x):
    return 2*x + 3 - 7

linear_solution = fsolve(
    linear_equation,
    0
)

print("Linear Equation Solution =", linear_solution)

# -------------------------
# Quadratic Equation
# x^2 - 5x + 6 = 0
# -------------------------

quadratic_coefficients = [1, -5, 6]

quadratic_roots = np.roots(
    quadratic_coefficients
)

print("Quadratic Roots =", quadratic_roots)

# -------------------------
# Two Linear Equations
# 2x + y = 5
# x - y = 1
# -------------------------

A = np.array([
    [2, 1],
    [1, -1]
])

B = np.array([5, 1])

solution = np.linalg.solve(A, B)

print("x =", solution[0])
print("y =", solution[1])

# -------------------------
# Logarithmic Function
# -------------------------

log_values = np.log([1, 2, 3, 4])

print("Log Values =", log_values)

# -------------------------
# Exponential Function
# -------------------------

exp_values = np.exp([1, 2, 3])

print("Exponential Values =", exp_values)
```
