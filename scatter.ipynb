```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Hours": [1,2,3,4,5],
    "Marks": [20,40,50,70,90],
    "Attendance": [60,70,75,80,95]
}

df = pd.DataFrame(data)

# Scatter plot
plt.scatter(
    df["Hours"],
    df["Marks"],
    label="Hours vs Marks"
)

plt.scatter(
    df["Hours"],
    df["Attendance"],
    label="Hours vs Attendance"
)

# Labels
plt.xlabel("Hours")

plt.ylabel("Values")

plt.title("Scatter Plot")

plt.legend()

plt.grid()

plt.show()
```
