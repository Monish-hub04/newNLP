```python
import numpy as np

# Create array
a = np.array([10, 20, 30, 40, 50])

# 2D array
b = np.array([
    [1, 2],
    [3, 4]
])

# Array properties
print("Shape =", a.shape)
print("Dimensions =", a.ndim)
print("Data Type =", a.dtype)

# Statistical functions
print("Mean =", np.mean(a))
print("Median =", np.median(a))
print("Standard Deviation =", np.std(a))
print("Variance =", np.var(a))
print("Minimum =", np.min(a))
print("Maximum =", np.max(a))
print("Sum =", np.sum(a))

# Reshape
print("Reshape")
print(a.reshape(5,1))

# Random values
print("Random Values")
print(np.random.rand(3))

# Matrix multiplication
print("Dot Product")
print(np.dot(a, a))

# Transpose
print("Transpose")
print(b.T)

# Unique values
print("Unique")
print(np.unique(a))

# Sorting
print("Sorted")
print(np.sort(a))
```
