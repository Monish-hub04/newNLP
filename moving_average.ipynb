```python
import pandas as pd

# Dataset
data = [10,12,14,16,18,20,22]

df = pd.DataFrame(data, columns=["Values"])

# SMA
df["SMA"] = df["Values"].rolling(
    window=3
).mean()

# CMA
df["CMA"] = df["Values"].expanding().mean()

# EMA
df["EMA"] = df["Values"].ewm(
    span=3
).mean()

print(df)
```
