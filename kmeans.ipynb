```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans

# Create dataset
X, y = make_blobs(
    n_samples=300,
    centers=3,
    random_state=42
)

# -------------------------
# Elbow Method
# -------------------------

wcss = []

for i in range(1, 11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42
    )

    kmeans.fit(X)

    wcss.append(kmeans.inertia_)

# Plot elbow graph
plt.plot(range(1,11), wcss)

plt.xlabel("Number of Clusters")

plt.ylabel("WCSS")

plt.title("Elbow Method")

plt.grid()

plt.show()

# -------------------------
# K-Means Clustering
# -------------------------

model = KMeans(
    n_clusters=3,
    random_state=42
)

# Train model
model.fit(X)

# Cluster labels
labels = model.predict(X)

# Cluster centers
centers = model.cluster_centers_

# Plot clusters
plt.scatter(
    X[:,0],
    X[:,1],
    c=labels
)

# Plot cluster centers
plt.scatter(
    centers[:,0],
    centers[:,1],
    s=300,
    c='red',
    marker='X'
)

plt.xlabel("Feature 1")

plt.ylabel("Feature 2")

plt.title("K-Means Clustering")

plt.grid()

plt.show()
```
