```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Department": ["IT", "CSE", "ECE", "MECH"],
    "Students": [50, 60, 40, 30]
}

df = pd.DataFrame(data)

# Bar plot
plt.bar(
    df["Department"],
    df["Students"]
)

# Labels
plt.xlabel("Department")

plt.ylabel("Students")

plt.title("Bar Plot")

plt.grid()

plt.show()
```
